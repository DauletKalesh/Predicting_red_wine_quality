<h2> <center> Regression </center> </h2>

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import  SVR
from sklearn.pipeline import Pipeline as PolynomialRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
regression_models = [
    LinearRegression(),
    Lasso(),
    Ridge(),
    PolynomialRegression((
        ("poly_features", PolynomialFeatures(degree = 4, include_bias=False)),
        ("lin_reg", LinearRegression()),
    )),
    SVR(),
    DecisionTreeRegressor(),
    KNeighborsRegressor(),
    RandomForestRegressor()
]
```

```python
features = ['volatile acidity', 'citric acid', 'total sulfur dioxide', 'sulphates', 'alcohol']
```

```python
X = new_wine[features]
y = new_wine['quality']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
[i.shape for i in [X_train, X_test, y_train, y_test]]
```

```python
import time
```

```python
fig, ax = plt.subplots(len(regression_models), 1, figsize = (6, 40))
for model in regression_models:
    start = time.time()
    model.fit(X_train, y_train)
    end = time.time()
    y_pred = model.predict(X_test)
    accuracies = cross_val_score(estimator = model, X = X_train, y = y_train, cv = 20)
    sns.regplot(y_test, y_pred, ax = ax[regression_models.index(model)])
    # ax[regression_models.index(model)].scatter(y_test, y_pred)
    ax[regression_models.index(model)].set_title(model.__class__.__name__)
    ax[regression_models.index(model)].axis([0, 10, 0, 10])
    ax[regression_models.index(model)].text(11, 9, f"MAE: {mean_absolute_error(y_test, y_pred)}")
    ax[regression_models.index(model)].text(11, 8, f"MSE: {mean_squared_error(y_test, y_pred)}")
    ax[regression_models.index(model)].text(11, 7, f"RMSE: {mean_squared_error(y_test, y_pred)**0.5}")
    ax[regression_models.index(model)].text(11, 6, f"R2_score: {r2_score(y_test, y_pred)}")
    ax[regression_models.index(model)].text(11, 5, f"{accuracies.mean() = }")
    ax[regression_models.index(model)].text(11, 4, f"{accuracies.std() = }")
    ax[regression_models.index(model)].text(10.5, 3, f"Training time: {end - start} seconds")
    ax[regression_models.index(model)].grid()

fig.show()
```

<h2> <center> Classification </center> </h2>

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import  SVC
from sklearn.ensemble import GradientBoostingClassifier
```

```python
classification_models = [
    LogisticRegression(),
    SVC(),
    DecisionTreeClassifier(random_state=42),
    KNeighborsClassifier(),
    RandomForestClassifier(random_state=42),
    GradientBoostingClassifier()
]
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
X = new_wine[features]
y = new_wine['quality_label']#.apply(lambda x: 1 if x >= 6.5 else 0)
# 1 -- good
# 0 -- poor
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
lst = []
for model in classification_models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    # print(model.__class__.__name__)
    report = pd.DataFrame(classification_report(y_test, y_pred, output_dict = True)).T
    # print(report)
    # print(classification_report(y_test, y_pred))
    cm = pd.DataFrame(confusion_matrix(y_test, y_pred))
    # print(cm)
    df = pd.concat( (cm, report), axis = 1, keys = ['confusion_matrix', 'classification report'] )
    lst.append(df)
    # print(df)
pd.concat(lst, axis = 0, keys = [i.__class__.__name__ for i in classification_models]).fillna('-')
```

### GridSearch for RandomForest

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [100, 200, 300, 500],
    'max_features': ['auto', None, 'log2']
}
clf = GridSearchCV(RandomForestClassifier(random_state=42), param_grid, cv=5, scoring='accuracy')
clf.fit(X_train, y_train)
clf.best_params_
```
