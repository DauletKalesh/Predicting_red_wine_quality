<h2> <center> Regression </center> </h2>

```python
from sklearn.linear_model import LinearRegression, LogisticRegression, Lasso, Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.svm import SVC, SVR
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
regression_models = [
    LinearRegression(),
    Lasso(),
    Ridge(),
    Pipeline((
        ("poly_features", PolynomialFeatures(degree=3, include_bias=False)),
        ("lin_reg", LinearRegression()),
    )),
    SVR(),
    DecisionTreeRegressor(),
    KNeighborsRegressor(),
    RandomForestRegressor()
]
```

```python
X = wine[
    ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'total sulfur dioxide', 'pH', 'sulphates', 'alcohol']
]
y = wine['quality']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
[i.shape for i in [X_train, X_test, y_train, y_test]]
```

```python
import time
```

```python
fig, ax = plt.subplots(len(regression_models), 1, figsize = (6, 40))
for model in regression_models:
    model.fit(X_train, y_train)
    start = time.time()
    y_pred = model.predict(X_test)
    end = time.time()
    ax[regression_models.index(model)].scatter(y_test, y_pred)
    ax[regression_models.index(model)].set_title(model.__class__.__name__)
    ax[regression_models.index(model)].axis([0, 10, 0, 10])
    ax[regression_models.index(model)].text(11, 9, f"MAE: {mean_absolute_error(y_test, y_pred)}")
    ax[regression_models.index(model)].text(11, 8, f"MSE: {mean_squared_error(y_test, y_pred)}")
    ax[regression_models.index(model)].text(11, 7, f"RMSE: {mean_squared_error(y_test, y_pred)**0.5}")
    ax[regression_models.index(model)].text(10.5, 6, f"Training time: {end - start} seconds")
    ax[regression_models.index(model)].grid()
    
```

<h2> <center> Classification </center> </h2>

```python
classification_models = [
    LogisticRegression(),
    SVC(),
    DecisionTreeClassifier(random_state=42),
    KNeighborsClassifier(),
    RandomForestClassifier(random_state=42)
]
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
X = wine[
    ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'total sulfur dioxide', 'pH', 'sulphates', 'alcohol']
]
y = wine['quality'].apply(lambda x: 1 if x >= 6.5 else 0)
# 1 -- good
# 0 -- poor
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
lst = []
for model in classification_models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    # print(model.__class__.__name__)
    report = pd.DataFrame(classification_report(y_test, y_pred, output_dict = True)).T
    # print(report)
    # print(classification_report(y_test, y_pred))
    cm = pd.DataFrame(confusion_matrix(y_test, y_pred))
    # print(cm)
    df = pd.concat( (cm, report), axis = 1, keys = ['confusion_matrix', 'classification report'] )
    lst.append(df)
    # print(df)
pd.concat(lst, axis = 0, keys = [i.__class__.__name__ for i in classification_models])
```
